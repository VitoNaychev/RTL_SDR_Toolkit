```python
import numpy as np
from fftsink import FftSink
from recordsamp import RecordSamp
import matplotlib.pyplot as plt
```

```python
samp_record = RecordSamp('adsb26.npy')
samp_rate = 1e6
samples = samp_record.load_from_file('adsb26.npy')
```

```python
fft_sink = FftSink(samp_rate, len(samples), y_lower_lim = -100)
fft_sink.update_fft(FftSink.calc_fft(samples, samp_rate, len(samples), average = True))
plt.show()
```

```python
F_offset = 70e3
s1 = samples * np.exp(1.0j*2.0*np.pi* F_offset/samp_rate*np.arange(len(samples)))  

fft_sink = FftSink(samp_rate, len(s1), y_lower_lim = -100)
fft_sink.update_fft(FftSink.calc_fft(s1, samp_rate, len(samples), average = True))
plt.show()
```
